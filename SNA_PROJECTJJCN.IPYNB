```python

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from networkx.algorithms import community
from itertools import cycle, islice
```

## Plotting the network

```python


# Ruta del archivo
file_path = 'out.maayan-faa'

# Crear un grafo vacío
G = nx.Graph()

# Leer el archivo y agregar las aristas al grafo
with open(file_path, 'r') as file:
    lines = file.readlines()

for line in lines:
    if line.startswith('%'):
        continue  # Omitir las líneas de comentarios
    nodes = line.strip().split()
    if len(nodes) == 2:
        G.add_edge(nodes[0], nodes[1])

# Dibujar la red
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=False, node_size=10, font_size=10)
plt.title('Network Visualization')
plt.show()

```

## Network characteristics

```python
size_of_network = G.number_of_nodes()
number_of_links = G.number_of_edges()

print(f"Size of network: {size_of_network}")
print(f"Number of link: {number_of_links}")
```

### Distance Metrics

```python
#Distance metrics 
diameter = nx.diameter(G)
eccentricity = nx.eccentricity(G)
radius = nx.radius(G)
periphery = list(nx.periphery(G))
center = list(nx.center(G))
density = nx.density(G)
transitivity = nx.transitivity(G)
average_path_length = nx.average_shortest_path_length(G)
average_clustering = nx.average_clustering(G)
print(f"Diameter: {diameter}")
print(f"Eccentricity: {eccentricity}")
print(f"Radius: {radius}")
print(f"Periphery: {periphery}")
print(f"Center: {center}")
print(f"Density: {density}")
print(f"Transitivity: {transitivity}")
print(f"Average clustering coefficient: {average_clustering}")
print(f"Average path length: {average_path_length}")
```

###  Degree distribution

```python
degree_distribution = [d for n, d in G.degree()]
plt.hist(degree_distribution)
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.bincount(degree_sequence)
degrees = np.arange(len(degree_count))
     

# Print degree sequence, degree count, and degrees
print("Degree Sequence:", degree_sequence)
print("Degree Count:", degree_count)
print("Degrees:", degrees)

plt.figure(figsize=(10, 6))
plt.bar(degrees, degree_count, width=0.80, color='b')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.show()
```

## Centrality Measures 

### Degree Centrality

```python
degree_centrality = nx.degree_centrality(G)
print(f"Degree centrality: {degree_centrality}")
```

### Betweenness centrality

```python
betweenness_centrality = nx.betweenness_centrality(G)
print(f"Betweenness centrality: {betweenness_centrality}")
```

### Closeness centrality

```python
closeness_centrality = nx.closeness_centrality(G)
print(f"Closeness centrality: {closeness_centrality}")
```

### Eigenvector centrality

```python
eigenvector_centrality = nx.eigenvector_centrality(G)
print(f"Eigenvector centrality: {eigenvector_centrality}")
```

### Average centrality measures

```python
#average centrality measures

average_degree_centrality = np.mean(list(degree_centrality.values()))
average_betweenness_centrality = np.mean(list(betweenness_centrality(G).values()))
average_closeness_centrality = np.mean(list(closeness_centrality.values()))
average_eigenvector_centrality = np.mean(list(eigenvector_centrality.values()))
```

```python
print(f"Average degree centrality: {average_degree_centrality}")
print(f"Average eigenvector centrality: {average_eigenvector_centrality}")
print(f"Average betweenness centrality: {average_betweenness_centrality}")
print(f"Average closeness centrality: {average_closeness_centrality}")
```

###  Plot of the centrality measures

```python
def plot_centrality(G, pos, centrality, title):
    plt.figure(figsize=(12, 8))
    nodes = nx.draw_networkx_nodes(G, pos, node_size=[v * 10000 for v in centrality.values()],
                                   node_color=list(centrality.values()), cmap=plt.cm.magma, alpha=0.8 )
    edges = nx.draw_networkx_edges(G, pos, alpha=0.3)
    plt.colorbar(nodes)
    plt.title(title)
    plt.show()
pos = nx.spring_layout(G)
```

```python
plot_centrality(G, pos, degree_centrality, 'Degree Centrality')
plot_centrality(G, pos, betweenness_centrality(G), 'Betweenness Centrality')
plot_centrality(G, pos, closeness_centrality, 'Closeness Centrality')
plot_centrality(G, pos, eigenvector_centrality, 'Eigenvector Centrality')
```

```python

degree_centrality = nx.degree_centrality(G) 
top_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]


print("Top 5 nodes by Degree Centrality:")
for i, (node, centrality) in enumerate(top_degree_centrality, start=1):
    print(f"{i}. Node: {node}, Degree Centrality: {centrality}")
```

```python
# plot the degree distribution in the network

# Get the top 5 nodes with highest degree centrality
top_nodes = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

# Extract the node labels and centrality values
labels = [node[0] for node in top_nodes]
centrality = [node[1] for node in top_nodes]

# Plot the top 5 nodes
plt.bar(labels, centrality)
plt.xlabel('Nodes')
plt.ylabel('Degree Centrality')
plt.title('Top 5 Nodes by Degree Centrality')
plt.show()
```

### Plotting the communities in pyvis

```python
from IPython.display import display, HTML
from pyvis.network import Network
import networkx as nx
from networkx.algorithms.community import greedy_modularity_communities
import matplotlib.colors as mcolors
import random

file_path = 'out.maayan-faa'  

# Crear el grafo
G = nx.Graph()

# Leer el archivo y agregar enlaces al grafo
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

for line in lines:
    if line.startswith('%'):
        continue  
    nodes = line.strip().split()
    if len(nodes) == 2:
        G.add_edge(nodes[0], nodes[1])

# Detectar comunidades
communities = list(greedy_modularity_communities(G))
print(f"Número de comunidades detectadas: {len(communities)}")

# Crear un diccionario para asignar a cada nodo su comunidad
node_community = {}
for i, community in enumerate(communities):
    for node in community:
        node_community[node] = i

# Obtener colores para las comunidades
colors = list(mcolors.CSS4_COLORS.keys())
random.shuffle(colors)

# Crear la visualización con pyvis
net = Network(notebook=True, cdn_resources='remote')

# Agregar nodos con colores de comunidad
for node, community in node_community.items():
    net.add_node(node, label=node, color=colors[community % len(colors)], title=f'Node {node} pertenece a la comunidad {community}')

# Agregar enlaces
for edge in G.edges():
    net.add_edge(edge[0], edge[1])

# Configurar opciones de visualización en 3D con nodos rápidos
net.set_options("""
var options = {
  "layout": {
    "improvedLayout": false,
    "hierarchical": false,
    "randomSeed": null
  },
  "physics": {
    "forceAtlas2Based": {
      "gravitationalConstant": -1000,
      "centralGravity": 0.01,
      "springLength": 100,
      "springConstant": 0.08,
      "damping": 0.4,
      "avoidOverlap": 0.1
    },
    "minVelocity": 1.0,
    "solver": "forceAtlas2Based",
    "timestep": 0.5,
    "stabilization": {
      "enabled": true,
      "iterations": 1000,
      "updateInterval": 25
    }
  },
  "interaction": {
    "hover": true,
    "tooltipDelay": 200,
    "zoomView": true,
    "dragView": true
  },
  "manipulation": {
    "enabled": true
  }
}
""")

# Mostrar la red en un archivo HTML
net.show("network.html")
display(HTML("network.html"))
```
